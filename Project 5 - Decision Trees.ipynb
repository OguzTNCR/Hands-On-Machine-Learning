```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier


iris = load_iris()

X = iris.data[ : , 2: ]
y = iris.target

```

```python
tree_classifier = DecisionTreeClassifier(max_depth = 2)
tree_classifier.fit(X, y)
```

```python
print(tree_classifier.predict_proba([[5, 1.5]]))
```

```python
tree_classifier.predict([[5, 1.5]])
```

```python
def plot_decision_boundary(classifier, X, y, axes, cmap):
    x1, x2 = np.meshgrid(np.linspace(axes[0], axes[1], 150),
                         np.linspace(axes[2], axes[3], 150))

    X_new = np.c_[x1.ravel(), x2.ravel()]
    y_pred = classifier.predict(X_new).reshape(x1.shape)

    plt.contourf(x1, x2, y_pred, alpha = 0.3, cmap = cmap)
    plt.contour(x1, x2, y_pred, alpha = 0.8, cmap = "Greys")

    colors = {"Wistia": ["#78785c", "#c47b27"], "Pastel1": ["red", "blue"]}
    markers = ("o", "^")
    for idx in (0, 1):
        plt.plot(X[:, 0][y == idx], X[:, 1][y == idx],
                 color=colors[cmap][idx], marker=markers[idx], linestyle="none")
```

```python
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=150, noise = 0.2, random_state=42)

tree1 = DecisionTreeClassifier(random_state=42)
tree2 = DecisionTreeClassifier(min_samples_leaf=5, random_state=42)

tree1.fit(X, y)
tree2.fit(X, y)
```

```python
plot_decision_boundary(tree1, X, y,
                       axes=[-1.5, 2.4, -1, 1.5], cmap="Wistia")
```

```python
plot_decision_boundary(tree2, X, y,
                       axes=[-1.5, 2.4, -1, 1.5], cmap="Wistia")
```
